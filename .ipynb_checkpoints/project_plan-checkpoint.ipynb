## Questions

Question 2: We would like to know which "kinds" of players are most likely to contribute a large amount of data so that we can target those players in our recruiting efforts.

Specific question: Can age predict the total amount of gameplay data a player contributes on the MineCraft server?

To address the specific question, we would use each player's age as the explanatory variable and their played hours as the response variable. By wrangling the data set to select these two variables for each participant, we can visualize a potential relationship and fit a simple linear regression model to determine if age is a meaningful factor in total amount of gameplay.

## Exploratory Data Analysis and Visualization

```r
# Loading data set into R
players <- read_csv("data/players.csv")
```

```r
# Minimum wrangling of data
players_select <- players |>
    select(Age, played_hours) |>
    drop_na()
players_select
```

```r
# Mean values for each variable in players data set
mean_table <- players_select |>
    summarise(mean_age = mean(Age, na.rm = TRUE), mean_played_hours = mean(played_hours, na.rm = TRUE))
mean_table
```

### Exploratory Visualizations

```r
# Distribution of player's ages
options(repr.plot.width = 6, repr.plot.height = 8)
age_distribution <- players_select |>
  ggplot(aes(x = Age)) +
  geom_histogram(binwidth = 2, fill = "skyblue", color = "black") +
  labs(title = "Distribution of Player Ages", x = "Age (years)", y = "Count of Players") +
  theme(text = element_text(size = 12))
age_distribution
```

```r
# Distribution of played hours per each player
options(repr.plot.width = 6, repr.plot.height = 8)
hours_distribution <- players_select |>
  ggplot(aes(x = played_hours)) +
  geom_histogram(binwidth = 10, fill = "lightgreen", color = "black") +
  labs(title = "Distribution of Total Gameplay Hours", x = "Total Hours Played on Server", y = "Number of Players") +
  theme(text = element_text(size = 12))
hours_distribution
```

```r
# Relationship between played hours and age
options(repr.plot.width = 6, repr.plot.height = 8)
players_plot <- players_select |>
    ggplot(aes(x = Age, y = played_hours)) +
    geom_point() +
    labs(title = "Relationship Between Age and Gameplay Hours", x = "Age (years)", y = "Total Hours Played") +
    theme(text = element_text(size = 12))
players_plot
```

#### Insights
- Hours played exhibits high variability, with many players contributing few hours and few players contributing many hours; presence of extreme outliers
- Age shows little variation with limited diversity; no even representation across lifespan
- When examining the scatterplot of Age versus played_hours, the points appear widely spread with no obvious visual structure

## Methods and Plan

A *simple linear regression* model would be most appropriate to address this question and explore whether age is a plausible influence on the total amount of gameplay. This method examines how a quantitative variable changes in relation to another, which aligns well with the specific question I chose.

Applying simple linear regression requires a few assumptions: a linear relationship between explanatory variables and the response variable, each observation is independent, the spread of the errors stays consistent, and the errors are roughly normally distributed. However, the method can be limited because it  can be influenced by outliers and missing data can still introduce bias.

Using metrics like RMSE/RMSPE or $R^2$ and comparing models with different predictor sets will help to compare and select the best model to address our question.

Before modelling, we will process the data by performing a single split into training and testing sets, where the training set involves 70% of the data and the leftover 30% will be tested, to prevent information leakage
All data pre-processing steps, like centering and scaling numeric variables and handling missing values, will be done within the training-set workflow only.

We will incorporate k-fold cross-validation on the training set to tune model specifications and evaluate stability, but the testing set will remain untouched until the final model is selected. This approach ensures fairness, prevents overfitting, and provides a principled framework for evaluating whether linear regression is suitable for answering the research question.
